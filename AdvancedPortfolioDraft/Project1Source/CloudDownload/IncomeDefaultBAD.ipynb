```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Loan_default.csv')

# Group data by 'Default' and calculate the average income for each group
average_income_by_default = df.groupby('Default')['Income'].mean().reset_index()

# Create a grouped bar chart
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = average_income_by_default['Default']

plt.bar(index, average_income_by_default['Income'], bar_width, label='Average Income', color='skyblue')
plt.xlabel('Default Status')
plt.ylabel('Average Income')
plt.title('Average Income by Default Status')
plt.xticks(index, ['No Default', 'Default'])
plt.legend()
plt.show()
```
