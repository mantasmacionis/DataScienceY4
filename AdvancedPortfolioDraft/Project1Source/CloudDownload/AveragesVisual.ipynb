```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Loan_default.csv')

# Calculate the average loan amount for each age
average_loan_by_age = df.groupby('Age')['LoanAmount'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(average_loan_by_age['Age'], average_loan_by_age['LoanAmount'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Average Loan Amount')
plt.title('Average Loan Amount by Age')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
