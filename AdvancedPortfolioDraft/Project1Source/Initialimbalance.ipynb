```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "Loan_default.csv"
df = pd.read_csv(file_path)

# Count the number of occurrences of each class in the "Default" column
default_counts = df['Default'].value_counts()

# Calculate the skewness of the "Default" column
skewness = default_counts[1] / default_counts[0]

# Define custom colors for the bars
colors = ['#FF6F61', '#6B5B95']

# Visualize the skewness using a bar chart with custom colors
plt.figure(figsize=(8, 6))
sns.barplot(x=default_counts.index, y=default_counts.values, palette=colors)
plt.title("Default Column Skewness")
plt.xlabel("Default")
plt.ylabel("Count")
plt.xticks([0, 1], ["No Default (0)", "Default (1)"])

# Add custom colors to the legend
color_labels = ['No Default', 'Default']
legend_labels = [plt.Line2D([0], [0], marker='o', color='w', label=label, markerfacecolor=color, markersize=15) for label, color in zip(color_labels, colors)]
plt.legend(legend_labels, color_labels)

plt.show()

# Print the skewness value
print(f"Skewness of Default column: {skewness:.2f}")
```
