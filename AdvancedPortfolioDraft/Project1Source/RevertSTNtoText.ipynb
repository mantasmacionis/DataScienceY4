```python
import pandas as pd

# Load the CSV dataset with numeric values
df = pd.read_csv("SMOTEofNumeric.csv")

# Define reverse mappings for each non-numeric column
reverse_education_mapping = {
    1: "Bachelor's",
    2: "High School",
    3: "PhD",
    4: "Master's",
    
}

reverse_employment_mapping = {
    1: "Full-time",
    2: "Part-time",
    3: "Unemployed",
    4: "Self-employed",
   
}

reverse_marital_status_mapping = {
    1: "Single",
    2: "Married",
    3: "Divorced",
    
}

reverse_has_mortgage_mapping = {
    0: "No",
    1: "Yes",
}

reverse_has_dependents_mapping = {
    0: "No",
    1: "Yes",
}

reverse_loan_purpose_mapping = {
    1: "Other",
    2: "Auto",
    3: "Business",
    4: "Education",
    5: "Home",
    
}

reverse_has_co_signer_mapping = {
    0: "No",
    1: "Yes",
}

# Apply the reverse mappings to the respective columns
df['Education'] = df['Education'].map(reverse_education_mapping)
df['EmploymentType'] = df['EmploymentType'].map(reverse_employment_mapping)
df['MaritalStatus'] = df['MaritalStatus'].map(reverse_marital_status_mapping)
df['HasMortgage'] = df['HasMortgage'].map(reverse_has_mortgage_mapping)
df['HasDependents'] = df['HasDependents'].map(reverse_has_dependents_mapping)
df['LoanPurpose'] = df['LoanPurpose'].map(reverse_loan_purpose_mapping)
df['HasCoSigner'] = df['HasCoSigner'].map(reverse_has_co_signer_mapping)

# Save the modified dataset with string values back to a new CSV file
df.to_csv("SMOTEFINAL.csv", index=False)

# Display the first few rows of the modified dataset
df.head()
```
